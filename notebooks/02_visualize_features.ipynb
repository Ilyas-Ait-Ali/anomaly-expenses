```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Load cleaned data
df = pd.read_csv('../data/creditcard_cleaned.csv')

# Check columns
print(df.columns.tolist())

# Remove target column from feature list
features = [col for col in df.columns if col != 'is_fraud']

# === HISTOGRAMS FOR ALL FEATURES ===
n = len(features)
cols = 4
rows = math.ceil(n / cols)

plt.figure(figsize=(20, 3 * rows))
for i, feature in enumerate(features):
    plt.subplot(rows, cols, i + 1)
    sns.histplot(data=df, x=feature, bins=50, kde=False)
    plt.title(f'Histogram of {feature}')
    plt.tight_layout()
plt.show()

# === BOXPLOTS FOR SELECTED FEATURES ===
subset = ['V1', 'V2', 'V10', 'Amount']

for feature in subset:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='is_fraud', y=feature, data=df)
    plt.title(f'Boxplot of {feature} by Fraud Status')
    plt.show()

# === COLORED HISTOGRAMS (OPTIONAL) ===
for feature in subset:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=df, x=feature, hue='is_fraud', bins=50, stat='density', common_norm=False)
    plt.title(f'Distribution of {feature} by Fraud Label')
    plt.show()
```
